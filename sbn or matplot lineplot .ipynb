```python
import matplotlib.pyplot as plt
import seaborn as sbn
import pandas as pd
```

```python
temp=[25,20,32,36,25]
city=['khi','lhr','isl','is','il']
```

```python
plt.plot(temp,city)

plt.grid(linewidth=1)

plt.show()
```

```python
temp=[25,20,32,36,25]
city=['khi','lhr','isl','is','il']
df=pd.DataFrame({'temperature':temp,
                'city':city})
```

```python
sbn.lineplot(x=temp,y=city,data=df,legend='auto')
```

# Scatter plot

```python
sales=pd.read_excel('Sample - Superstore Sales.xls')
```

```python
pd.set_option('display.max_columns',None)
```

```python
sales.head()
```

```python
sbn.relplot("Sales","Profit",data=sales)
```

```python
sbn.relplot('Sales','Profit',data=sales,hue='Order Priority')
```

```python
sbn.relplot('Sales','Profit',data=sales,hue='Order Priority',style='Order Priority')
plt.show()
```

```python
sbn.relplot('Sales','Profit',data=sales,size='Discount',hue='Product Base Margin',sizes=(30,200))
plt.show()
```

```python
sbn.relplot('Order Date','Sales',kind='line',data=sales)
```

```python
sales['year']=sales['Order Date'].dt.year
```

```python
sales['year'].unique()
```

```python
sbn.relplot('year','Sales',data=sales,kind='line')
```

```python
sbn.relplot('year','Sales',data=sales,kind='line',estimator=None)
```

```python
sbn.relplot(x='year',y='Sales',data=sales,kind='line',hue='Product Category',col='Region',col_wrap=3)
```

```python
sbn.relplot(x='year',y='Sales',data=sales,kind='line',hue='Product Category',col='Region')
```

```python
sbn.relplot(x='year',y='Sales',data=sales,kind='line',estimator=None,hue='Ship Mode')
```

```python
sbn.relplot(x='Profit',y='Sales',data=sales,col='Customer Segment',row='Product Category',hue='Ship Mode')
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn
import pandas as pd
```

```python
temp=[25,20,32,36,25]
city=['khi','lhr','isl','is','il']
```

```python
plt.plot(temp,city)
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.grid(linewidth=1)

plt.show()
```
