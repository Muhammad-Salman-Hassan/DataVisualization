```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Sample - Superstore Sales.xls')
df.head(3)
```

```python
sbn.catplot(x="Product Category",y='Shipping Cost',data=df,kind='box',hue='Ship Mode')
```

```python
sbn.catplot(x="Product Category",y='Shipping Cost',data=df,kind='box',hue='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air'])
```

```python
sbn.catplot(x="Product Category",y='Shipping Cost',data=df,kind='box',hue='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air']
            ,row='Ship Mode')
```

```python
sbn.catplot(x="Product Category",y='Shipping Cost',data=df,kind='box',hue='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air'],
           col='Ship Mode')
```

```python
sbn.catplot(x="Product Category",y='Shipping Cost',data=df,kind='box',hue='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air'],
           col='Ship Mode',row='Customer Segment')
```
